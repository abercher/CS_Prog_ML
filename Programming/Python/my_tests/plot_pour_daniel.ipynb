```python
import pylab
import numpy

x = numpy.linspace(0.1, 1.43, 100) # 100 linearly spaced numbers
y = (2.63/100 + (45/100*(1.43 - x)))*100

# compose plot
pylab.plot(x,y)
pylab.show() # show the plot
pylab.savefig('interet_courbe.png')
```
